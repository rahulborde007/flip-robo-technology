```python
# Display columns with null values
columns_with_null = null_values[null_values > 0].index
print("Columns with null values:", columns_with_null)
```

```python
# Fill null values with the mode of each column
for column in columns_with_null:
    mode_value = stu_data[column].mode()[0]
    stu_data[column].fillna(mode_value, inplace=True)

# Check again for null values to confirm they have been filled
print(stu_data.isnull().sum())
```

```python
categorical_columns = stu_data.select_dtypes(include=['object']).columns

# One-hot encode the categorical columns
df_encoded = pd.get_dummies(stu_data, columns=categorical_columns)

print(df_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('CGPA', axis=1)  # Features
y = df_encoded['CGPA']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=41)
```

```python
print("X-train shape:",X_train.shape)
print("X-test Shape:",X_test.shape)
print("y-train Shape:",y_train.shape)
print("y-test Shape:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
#  Data Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score, KFold
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Initialize different regression models
models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    RandomForestRegressor(),
    SVR()
]

# Set up a pipeline with data standardization for each model
pipelines = [make_pipeline(StandardScaler(), model) for model in models]

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Iterate over models and evaluate using cross-validation
for model, pipeline in zip(models, pipelines):
    cross_val_scores = cross_val_score(pipeline, X, y, cv=kf, scoring='r2')
    mean_cv_score = cross_val_scores.mean()
    print(f"{model.__class__.__name__}: Mean Cross-Validation Score: {mean_cv_score:.2f}")
```

```python
for model, pipeline in zip(models, pipelines):
    
    pipeline.fit(X_train, y_train)
    
    # Predict on the test data
    y_pred = pipeline.predict(X_test)
    
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.title('Actual vs. Predicted CGPA')
plt.xlabel('Actual CGPA')
plt.ylabel('Predicted CGPA')
plt.show()
```

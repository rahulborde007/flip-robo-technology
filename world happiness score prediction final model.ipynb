```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
Happi_data.head()
```

```python
# lets drop the target 
```

```python
X = Happi_data.drop(['Happiness Score'], axis=1)
y = Happi_data['Happiness Score']
```

```python
# split the  data in train and test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 42)
```

```python
print("X-train shape:",X_train.shape)
print("X-test Shape:",X_test.shape)
print("y-train Shape:",y_train.shape)
print("y-test Shape:",y_test.shape)
```

```python
# standardization of data
# model building (regression model)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    score = model.score(X_test,y_test)
    
    score_percentage = score * 100
    
    print ('{} Score: {:.2f}%'.format(type(model).__name__, score_percentage))
    
lr = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
ada = AdaBoostRegressor()
knn = KNeighborsRegressor()

model(lr,X_train, y_train, X_test, y_test)

model(dt,X_train, y_train, X_test, y_test)

model(rf,X_train, y_train, X_test, y_test)

model(ada,X_train, y_train, X_test, y_test)

model(knn,X_train, y_train, X_test, y_test)
    
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    print('{} Metrics:'.format(type(model).__name__))
    print('Mean Squared Error: {:.2f}'.format(mse))
    print('R-squared (R2) Score: {:.2f}'.format(r2))
    print('Mean Absolute Error: {:.2f}'.format(mae))
    print()
    
models = [LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(),AdaBoostRegressor(),KNeighborsRegressor()]

for model in models:
    evaluate_model(model,X_train, y_train, X_test, y_test)
```
